```python
def triangularNumber(n):
    tN = int(n*(n+1)/2)
    return tN

def makeCoeff(ci):
    X = np.array([])
    for row in ci:
        X = np.append(X, row**2)
        combs = combinations(list(row), 2)
        num_rows = np.shape(ci)[0] 
        for comb in combs:
            X = np.append(X, comb[0]*comb[1]) 
    
    X = X.reshape(num_rows, triangularNumber(len(ci[0])))
    
    return X
```
